```python
# Import webclient, libraries
! pip install chembl-webresource-client

import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# Target search for coronavirus
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
# Select entry [6] SARS Coronavirus Single Protein
selected_target = targets.target_chembl_id[6]
selected_target
```

```python
# Retrieve bioactivity data from selected target
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
pd.set_option('display.max_rows', 10)
```

```python
df = pd.DataFrame.from_dict(res)
df
```

```python
# Export data to csv
df.to_csv('bioactivity_data.csv', index=False)
```

```python
# Cleaning data: dropping compounds with missing entries for standard_value
df2 = df[df.standard_value.notna()]
df2
```

```python
# Data Preprocessing
df2.molecule_chembl_id
```

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2[selection]
df3
```

```python
df3.to_csv('sarscovid_02_bioactivity_data_preprocessed.csv', index=False)
```

```python
df4 = pd.read_csv('sarscovid_02_bioactivity_data_preprocessed.csv')
```

```python
bioactivity_threshold = []
for i in df4.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
```

```python
bioactivity_class = pd.Series(bioactivity_threshold, name='class')
df5 = pd.concat([df4, bioactivity_class], axis=1)
df5
```

```python
df5.to_csv('sarscovid_03_bioactivity_data_curated.csv', index=False)
```
